```python
def getSmoothData(data_index, window_size, end_size):
    out = []
    
    sum = 0
    for i in tqdm(range(0, window_size)):
        sum += abs(df[data_index][i])
    sum /= window_size
    
    for i in tqdm(range(window_size, end_size)):    
        sum *= window_size
        sum += abs(df[data_index][i])
        sum -= abs(df[data_index][i - window_size])
        sum /= window_size

        out.append(sum)
        
    return out
```

```python
l1 = getSmoothData(1, 10000, 6000000)
```

```python
def getmids(data, end_size, threshold):
    mids = []
    start = 0
    end = 0

    for i in tqdm(range(0, end_size)):
        value = data[i]

        if (value < threshold):
            if (start == 0):
                start = i
            end = i

        if (value > threshold):
            if (start != 0):
                mid = (int)((start + end) / 2)
                mids.append(mid)

                start = 0

    return mids
```

```python
def deduplication(datas):
    pre = datas[0]
    for data in datas[1:]:
        if (data - pre < 10000):
            mids.remove(data) 
        else:
            pre = data
            
    return datas
```

```python
def split(datas, mids):
    idx_pre = 0
    idx_now = 0
    out = []
    for data in mids:
        idx_now = data
        out.append(datas[idx_pre:idx_now])
        idx_pre = idx_now
        
    return out
```
